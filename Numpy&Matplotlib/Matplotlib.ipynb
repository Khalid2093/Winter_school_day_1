```python
import matplotlib.pyplot as plt
```

Line Chart

```python
x = [1, 2, 3, 4, 5]
y = [2, 5, 6, 20, 14]

plt.plot(x, y)
```

Bar Chart

```python
categories = ['Category A', 'Category B', 'Category C']
values = [4, 7, 2]

plt.bar(categories, values)
```

Histogram

```python
data = [2,2,2,3,10,10,10,9,9,8,5]
plt.hist(data, bins=9)
```

Scatter Plot

```python
x = [1, 2, 3, 4, 5]
y = [2, 14, 6, 18, 10]

plt.scatter(x, y)
```

Pie Chart

```python
labels = ['Label 1', 'Label 2', 'Label 3']
sizes = [30, 40, 30]

plt.pie(sizes, labels=labels)
```

Area Plot

```python
# Data
x = [1, 2, 3, 4, 5]
y1, y2 = [10, 20, 15, 25, 30], [5, 15, 10, 20, 25]

# Area Chart
plt.fill_between(x, y1, y2, color='skyblue', alpha=0.4, label='Area 1-2')
plt.plot(x, y1, label='Line 1', marker='o')
plt.plot(x, y2, label='Line 2', marker='x')

# Labels and Title
plt.xlabel('X-axis'), plt.ylabel('Y-axis'), plt.title('Area Chart Example')

# Legend and Display
plt.legend()
```
